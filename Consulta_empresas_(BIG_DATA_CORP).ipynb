```python
import re
import json
import requests

# üîê Credenciais
bigdata_token_id = '68375c090347ebbb5f2a6797'
bigdata_token_hash = 'eyJhbGciOiJIUzI1NiIsInR5cCI6IkpXVCJ9.eyJuYW1lIjoiR0VPUkdFLkZFSVRPU0FAQ0xPVURXQUxLLklPIiwianRpIjoiYjQ1NTJmOTQtNGFmNi00MThlLWIzM2EtNmM5MTE5ZTZiOGY5IiwibmFtZVVzZXIiOiJHZW9yZ2UgRmVpdG9zYSIsInVuaXF1ZV9uYW1lIjoiR0VPUkdFLkZFSVRPU0FAQ0xPVURXQUxLLklPIiwiZG9tYWluIjoiQ0xPVURXQUxLIiwicHJvZHVjdHMiOlsiQklHQk9PU1QiLCJCSUdJRCJdLCJuYmYiOjE3NDg0NTg1MDUsImV4cCI6MTc3OTk5NDUwNSwiaWF0IjoxNzQ4NDU4NTA1LCJpc3MiOiJCaWcgRGF0YSBDb3JwLiJ9.UZERauMBKE62U2sr7vWzBbRD0j79Eid0DKHSQkByf9I'

# üß† Fun√ß√£o de consulta √† API
def fetch_bdc_data(document_number, url, dataset, token_hash, token_id):
    headers = {
        "Accept": "application/json",
        "Content-Type": "application/json",
        "AccessToken": token_hash,
        "TokenId": token_id
    }
    payload = {
        "q": f"doc{{{document_number}}}",
        "Datasets": dataset
    }
    response = requests.post(url, json=payload, headers=headers)
    response.raise_for_status()
    return response.json()




# ‚úÖ Lista de CNPJs para consulta
cnpjs = [

"43728245000142"
]




# üî¢ Sanitiza√ß√£o dos CNPJs (caso venham formatados)
cnpjs_sanitizados = [re.sub(r'\D', '', cnpj) for cnpj in cnpjs]

# üöÄ Consulta em massa
resultados = []
for cnpj in cnpjs_sanitizados:
    print(f"üîç Buscando dados para: {cnpj}")
    try:
        bdc_data = fetch_bdc_data(
            document_number=cnpj,
            url="https://plataforma.bigdatacorp.com.br/empresas",
            dataset="""basic_data,
               processes.filter(partypolarity = PASSIVE),
               kyc.filter(standardized_type, standardized_sanction_type, type, sanctions_source = Conselho Nacional de Justi√ßa)""",
            token_hash=bigdata_token_hash,
            token_id=bigdata_token_id
        )
        resultados.append(bdc_data)
    except Exception as e:
        print(f"‚ö†Ô∏è Erro ao buscar dados para {cnpj}: {str(e)}")
        resultados.append({
            "document_number": cnpj,
            "error": str(e)
        })

# üìú Resultado final em JSON
resultado_json = json.dumps(resultados, indent=4, ensure_ascii=False)
print("\nüì¶ Resultado da consulta em massa:")
print(resultado_json)
```

```python
# Instalar depend√™ncia para salvar como Excel
!pip install openpyxl

import pandas as pd

# Processar os dados da vari√°vel `resultados`
records = []

for item in resultados:
    # Verifica se houve erro na consulta
    if "Result" not in item or not item["Result"]:
        continue

    for result in item["Result"]:
        bd = result.get("BasicData", {})
        main_activity = ""
        if "Activities" in bd:
            main = next((a for a in bd["Activities"] if a.get("IsMain")), {})
            main_activity = main.get("Activity", "")

        records.append({
            "CNPJ": bd.get("TaxIdNumber"),
            "Raz√£o Social": bd.get("OfficialName"),
            "Nome Fantasia": bd.get("TradeName", ""),
            "UF": bd.get("HeadquarterState"),
            "Situa√ß√£o": bd.get("TaxIdStatus"),
            "Regime Tribut√°rio": bd.get("TaxRegime"),
            "Capital (R$)": float(bd.get("AdditionalOutputData", {}).get("CapitalRS", 0.0)),
            "Data de Funda√ß√£o": bd.get("FoundedDate", "")[:10],
            "Atividade Principal": main_activity
        })

# Criar DataFrame
df = pd.DataFrame(records)

# Exibir preview
df.head()

# Salvar como Excel (.xlsx)
excel_filename = "resultado_cnpjs.xlsx"
df.to_excel(excel_filename, index=False)

# Oferecer download no Colab
from google.colab import files
files.download(excel_filename)
```
